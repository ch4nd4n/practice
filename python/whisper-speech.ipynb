```python
import whisper
model = whisper.load_model("base")
```

```python
speech_file = "/Users/chandan/Desktop/microservices.mp3"
transcriber = model.transcribe(speech_file,fp16=False, language='en', max_initial_timestamp=None)
```

```python
import datetime
```

```python
count = 1
text_file = open("caption.txt", "w")
for item in transcriber['segments']:
    text_file.write(str(count))
    text_file.write('\n')
    line = f"{datetime.timedelta(seconds=item['start'])} --> {datetime.timedelta(seconds=item['end'])}"
    text_file.write(line)
    text_file.write(item['text'])
    count = count + 1
    text_file.write('\n')
text_file.close()
```
