```python
header = ['Folio', 'Scheme', 'Close Calculated', 'RTA', 'NAV', 'Value', 'ISIN']
with open(outfile, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(header)
    for row in data.folios:
        print(row)
        csv_row = [row.folio, row.schemes[0].scheme.replace('\n', ' '), row.schemes[0].close_calculated, row.schemes[0].rta, row.schemes[0].valuation.nav, row.schemes[0].valuation.value, row.schemes[0].isin]

        writer.writerow(csv_row)
```
